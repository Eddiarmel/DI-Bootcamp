```python
from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]

print("Moyenne   :", stats.tmean(data))     # 16.625
print("Médiane   :", stats.scoreatpercentile(data, 50))  # 16.5
print("Variance  :", stats.tvar(data))      # 17.125  (ddof=1 par défaut)
print("Écart-type:", stats.tstd(data))      # 4.138   (ddof=1 par défaut)
```

Exercice 3 — Distribution normale tracée

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 50, 10
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 300)
pdf = norm.pdf(x, loc=mu, scale=sigma)

plt.figure(figsize=(8, 4))
plt.plot(x, pdf, color='steelblue', linewidth=2)
plt.fill_between(x, pdf, alpha=0.15, color='steelblue')
plt.axvline(mu, color='red', linestyle='--', label='μ = 50')
plt.title('Distribution normale  μ=50, σ=10')
plt.xlabel('x')
plt.ylabel('Densité de probabilité')
plt.legend()
plt.show()
```

Exercice 4 — Test T

```python
import numpy as np
from scipy import stats

np.random.seed(42)  # pour la reproductibilité
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_value = stats.ttest_ind(data1, data2)
print(f"T-statistique : {t_stat:.4f}")   # -8.5515
print(f"P-value       : {p_value:.6f}")  # ≈ 0.0

if p_value < 0.05:
    print("→ Différence SIGNIFICATIVE")
```

Exercice 5 — Régression linéaire immobilière

```python
import numpy as np
from scipy import stats

house_sizes  = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

print(f"Pente            : {slope:.2f}")          # 1828.77 €/m²
print(f"Ordonnée origine : {intercept:.2f}")       # 64383.56 €
print(f"R²               : {r_value**2:.4f}")      # 0.9884
print(f"Prix estimé 90m² : {slope*90 + intercept:.0f} €")  # 228 972 €
```

Exercice 6 — ANOVA engrais

```python
from scipy import stats

f1 = [5, 6, 7, 6, 5]
f2 = [7, 8, 7, 9, 8]
f3 = [4, 5, 4, 3, 4]

F, p = stats.f_oneway(f1, f2, f3)
print(f"F-value : {F:.4f}")    # 28.5263
print(f"P-value : {p:.8f}")    # 0.00002754
```

Exercice 7 ★ — Distribution binomiale

```python
from scipy import stats

binom_dist = stats.binom(n=10, p=0.5)
print(f"P(X=5) = {binom_dist.pmf(5):.4f}")  # 0.2461
```

Exercice 8 ★ — Corrélations

```python
import pandas as pd
from scipy import stats

data = pd.DataFrame({
    'age':    [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

r_p, p_p = stats.pearsonr(data['age'], data['income'])
r_s, p_s = stats.spearmanr(data['age'], data['income'])

print(f"Pearson  r = {r_p:.4f}  | p = {p_p:.6f}")   # r = 0.9997
print(f"Spearman r = {r_s:.4f}  | p = {p_s:.6f}")   # r = 1.0000
```
