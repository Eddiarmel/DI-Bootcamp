Tâche : Expliquer pourquoi la visualisation des données est importante dans l'analyse des données.
La visualisation des données est un outil essentiel du processus d'analyse. Elle permet de transformer des données brutes en représentations visuelles pertinentes, facilitant ainsi l'identification des tendances, des schémas et des anomalies
Tâche : Décrire l'utilité d'un graphique linéaire en visualisation de données.
Un graphique linéaire est particulièrement utile pour visualiser des données continues sur une période de temps. Il permet de montrer les tendances et les variations dans les données, ce qui peut aider à identifier des modèles ou des relations entre les variables. Par exemple, un graphique linéaire peut être utilisé pour suivre l'évolution des ventes mensuelles d'une entreprise ou pour observer les changements de température au fil du temps.

 Exercice 4 : Visualisation de la distribution des moyennes cumulatives

```python
import seaborn as sns
import pandas as pd
# Chargement du fichier CSV (ajustez le chemin si nécessaire)
df = pd.read_csv("Student Mental health.csv")

# Aperçu rapide des données
print("Aperçu du DataFrame :")
print(df.head())
print("\nInformations sur les colonnes :")
print(df.info())
print("\nValeurs uniques dans la colonne 'What is your CGPA?':")
print(df['What is your CGPA?'].unique())

# Création de l'histogramme des catégories de CGPA
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='What is your CGPA?', color='skyblue', edgecolor='black')

# Personnalisation du titre et des axes
plt.title("Distribution des catégories de CGPA parmi les étudiants", fontsize=14)
plt.xlabel("Plage de CGPA")
plt.ylabel("Nombre d'étudiants")
plt.xticks(rotation=45)

# Affichage du graphique
plt.tight_layout()
plt.show()
```


# EXERCICE 5 : Anxiété selon le sexe

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Student Mental health.csv")

plt.figure(figsize=(7, 5))

sns.countplot(
    data=df,
    x="Choose your gender",
    hue="Do you have Anxiety?",
    palette={"Yes": "salmon", "No": "steelblue"}
)

plt.title("Comparaison des niveaux d'anxiété selon le sexe")
plt.xlabel("Sexe")
plt.ylabel("Nombre d'étudiants")
plt.legend(title="Anxiété")
plt.tight_layout()
plt.show()
```


# EXERCICE 6 : Âge et crises de panique

```python

# Convertir les réponses en valeurs numériques : Yes=1, No=0
df["Panic_numeric"] = df["Do you have Panic attack?"].map({"Yes": 1, "No": 0})

# Supprimer les lignes avec Age manquant
df_clean = df.dropna(subset=["Age"])

plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=df_clean,
    x="Age",
    y="Panic_numeric",
    hue="Do you have Panic attack?",
    palette={"Yes": "tomato", "No": "steelblue"},
    s=100,
    alpha=0.6
    # jitter supprimé ← n'existe pas dans scatterplot
)

plt.title("Relation entre l'âge et les crises de panique")
plt.xlabel("Âge")
plt.ylabel("Crise de panique (0 = Non, 1 = Oui)")
plt.yticks([0, 1], ["Non (0)", "Oui (1)"])
plt.legend(title="Crise de panique")
plt.tight_layout()
plt.show()
```
